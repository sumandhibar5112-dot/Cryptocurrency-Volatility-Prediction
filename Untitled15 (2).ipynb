```python
from google.colab import files

uploaded = files.upload()
import pandas as pd

# Load the dataset
df = pd.read_csv('dataset (1).csv')

# Basic inspection
print("Data Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nFirst 5 rows:")
print(df.head())
print("\nData Info:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())
print("\nUnique Cryptocurrencies:", df['crypto_name'].nunique())
print("\nDate Range:", df['date'].min(), "to", df['date'].max())
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Convert date to datetime
df['date'] = pd.to_datetime(df['date'])
df = df.sort_values(['crypto_name', 'date'])

# Feature Engineering function
def engineer_features(group):
    # Log Returns
    group['log_return'] = np.log(group['close'] / group['close'].shift(1))

    # Target: 7-day rolling volatility (std of log returns)
    # We want to predict tomorrow's volatility, so we can shift the target or use features to predict current.
    # Assignment says "predict cryptocurrency volatility levels". Let's predict the volatility of the NEXT 7 days or just the rolling vol.
    group['volatility_7d'] = group['log_return'].rolling(window=7).std()

    # Technical Indicators
    group['range'] = (group['high'] - group['low']) / group['open']
    group['volume_change'] = group['volume'].pct_change()
    group['close_ma7'] = group['close'].rolling(window=7).mean()
    group['close_ma21'] = group['close'].rolling(window=21).mean()

    # Lagged Volatility (Very important for prediction)
    group['lag_vol_1'] = group['volatility_7d'].shift(1)

    # Target: Volatility 1 day ahead (what we want to predict)
    group['target_vol'] = group['volatility_7d'].shift(-1)

    return group

# Apply feature engineering per crypto
df = df.groupby('crypto_name', group_keys=False).apply(engineer_features)

# Drop rows with NaN (due to rolling/shifting)
df_clean = df.dropna().copy()

# EDA Visualizations
plt.figure(figsize=(15, 6))
for crypto in ['Bitcoin', 'Ethereum', 'Litecoin']:
    subset = df_clean[df_clean['crypto_name'] == crypto]
    plt.plot(subset['date'], subset['volatility_7d'], label=crypto)
plt.title('7-Day Rolling Volatility for Top Cryptocurrencies')
plt.xlabel('Date')
plt.ylabel('Volatility (Std Dev of Log Returns)')
plt.legend()
plt.savefig('volatility_trends.png')

plt.figure(figsize=(10, 8))
corr = df_clean[['close', 'volume', 'marketCap', 'log_return', 'volatility_7d', 'range', 'volume_change']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.savefig('correlation_matrix.png')

# Print summary
print("Final Dataset Shape after cleaning:", df_clean.shape)
print(df_clean[['crypto_name', 'date', 'volatility_7d', 'target_vol']].head())
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.preprocessing import LabelEncoder

# Encode crypto names (though we might drop it or use as feature)
le = LabelEncoder()
df_clean['crypto_id'] = le.fit_transform(df_clean['crypto_name'])

# Features and Target
features = ['open', 'high', 'low', 'close', 'volume', 'marketCap',
            'log_return', 'volatility_7d', 'range', 'volume_change',
            'close_ma7', 'close_ma21', 'lag_vol_1', 'crypto_id']
target = 'target_vol'

# Time-based split: Use data before 2022 for training, 2022 for testing
train_df = df_clean[df_clean['date'] < '2022-01-01']
test_df = df_clean[df_clean['date'] >= '2022-01-01']

X_train, y_train = train_df[features], train_df[target]
X_test, y_test = test_df[features], test_df[target]

# Replace any inf/-inf with 0 (sometimes happens with pct_change)
X_train = X_train.replace([np.inf, -np.inf], 0)
X_test = X_test.replace([np.inf, -np.inf], 0)

# Train Model
model = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
model.fit(X_train, y_train)

# Predictions
y_pred = model.predict(X_test)

# Metrics
mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae}")
print(f"RMSE: {rmse}")
print(f"R2 Score: {r2}")

# Plot Prediction vs Actual for a specific coin in test set (e.g., Bitcoin)
btc_test = test_df[test_df['crypto_name'] == 'Bitcoin'].copy()
btc_test['pred_vol'] = model.predict(btc_test[features].replace([np.inf, -np.inf], 0))

plt.figure(figsize=(12, 6))
plt.plot(btc_test['date'], btc_test['target_vol'], label='Actual Volatility')
plt.plot(btc_test['date'], btc_test['pred_vol'], label='Predicted Volatility', linestyle='--')
plt.title('Bitcoin Volatility Prediction (Test Set 2022)')
plt.legend()
plt.savefig('prediction_plot.png')

# Feature Importance
importances = model.feature_importances_
feat_importance = pd.Series(importances, index=features).sort_values(ascending=False)
plt.figure(figsize=(10, 6))
feat_importance.plot(kind='bar')
plt.title('Feature Importance')
plt.savefig('feature_importance.png')
```

```python
# Save the processed data with predictions for the test set
processed_data_output = df_clean.copy()
# Just for the test set visualization in the final CSV, let's add predictions where possible
processed_data_output['predicted_vol'] = np.nan
processed_data_output.loc[test_df.index, 'predicted_vol'] = y_pred

processed_data_output.to_csv('processed_cryptocurrency_data.csv', index=False)
print("Processed data saved to processed_cryptocurrency_data.csv")
```

1. Project Overview & Objective
The goal is to build a machine learning model to predict cryptocurrency volatility levels (e.g., Low, Medium, High) based on historical market data (OHLC, Volume, Market Cap). This helps traders and institutions anticipate market instability.

2. Solution Architecture
Input: Historical daily data (Open, High, Low, Close, Volume).

Preprocessing: Cleaning, sorting by date, calculating returns.

Feature Engineering: Rolling volatility, Moving Averages, Lag features.

Target: Next period's volatility level (Classification: Low, Medium, High).

Model: Random Forest Classifier.

Output: Predicted volatility level for the next period.

3. Implementation (Python Code)
The following code performs Data Processing, Feature Engineering, EDA, and Model Training as required by the assignment deliverables.

```python
from google.colab import files

uploaded = files.upload()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.preprocessing import LabelEncoder

# ==========================================
# 1. Data Loading & Preprocessing
# ==========================================
# Load the dataset
df = pd.read_csv('dataset.csv')

# Convert date to datetime and sort
df['date'] = pd.to_datetime(df['date'])
df.sort_values(by=['crypto_name', 'date'], inplace=True)

# Check for missing values
print("Missing values before cleaning:\n", df.isnull().sum())
df.dropna(inplace=True)

# Focus on Bitcoin for this analysis (or you can run a loop for all)
# We will create a model specifically for Bitcoin as a primary example
btc_df = df[df['crypto_name'] == 'Bitcoin'].copy()

# ==========================================
# 2. Feature Engineering
# ==========================================
# Calculate Daily Log Returns
btc_df['log_return'] = np.log(btc_df['close'] / btc_df['close'].shift(1))

# Calculate Volatility (7-day rolling standard deviation of returns)
window_size = 7
btc_df['volatility'] = btc_df['log_return'].rolling(window=window_size).std()

# Create Features for Prediction
# 1. Lagged Volatility (Volatility of the past few days)
btc_df['vol_lag1'] = btc_df['volatility'].shift(1)
btc_df['vol_lag7'] = btc_df['volatility'].shift(7)

# 2. Price Momentum (Returns)
btc_df['return_lag1'] = btc_df['log_return'].shift(1)

# 3. Volume Change
btc_df['volume_change'] = btc_df['volume'].pct_change()

# 4. Moving Averages
btc_df['SMA_7'] = btc_df['close'].rolling(window=7).mean()
btc_df['SMA_30'] = btc_df['close'].rolling(window=30).mean()

# Define Target Variable: Next Day's Volatility Level
# We want to predict if the volatility *tomorrow* will be Low, Medium, or High
btc_df['target_volatility'] = btc_df['volatility'].shift(-1) # Next day's volatility

# Drop NaNs created by rolling/shifting
btc_df.dropna(inplace=True)

# Discretize the target into levels (Low, Medium, High) using quantiles
# Low: Bottom 33%, Medium: Middle 33%, High: Top 33%
q33 = btc_df['target_volatility'].quantile(0.33)
q66 = btc_df['target_volatility'].quantile(0.66)

def categorize_volatility(val):
    if val < q33:
        return 'Low'
    elif val < q66:
        return 'Medium'
    else:
        return 'High'

btc_df['volatility_level'] = btc_df['target_volatility'].apply(categorize_volatility)

# ==========================================
# 3. Exploratory Data Analysis (EDA)
# ==========================================
plt.figure(figsize=(14, 7))

# Plot 1: Closing Price
plt.subplot(2, 2, 1)
plt.plot(btc_df['date'], btc_df['close'], label='Close Price', color='blue')
plt.title('Bitcoin Closing Price Over Time')
plt.legend()

# Plot 2: Volatility
plt.subplot(2, 2, 2)
plt.plot(btc_df['date'], btc_df['volatility'], label='7-Day Volatility', color='orange')
plt.title('Bitcoin Historical Volatility (7-Day Rolling)')
plt.legend()

# Plot 3: Distribution of Volatility Levels
plt.subplot(2, 2, 3)
sns.countplot(x='volatility_level', data=btc_df, order=['Low', 'Medium', 'High'])
plt.title('Distribution of Target Volatility Levels')

# Plot 4: Correlation Heatmap
plt.subplot(2, 2, 4)
features = ['vol_lag1', 'vol_lag7', 'return_lag1', 'volume_change', 'SMA_7', 'SMA_30']
sns.heatmap(btc_df[features].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Feature Correlation Matrix')

plt.tight_layout()
plt.show()

# ==========================================
# 4. Model Building
# ==========================================
# Select Features and Target
X = btc_df[['vol_lag1', 'vol_lag7', 'return_lag1', 'volume_change', 'SMA_7', 'SMA_30']]
y = btc_df['volatility_level']

# Split Data (Time-series split prevents data leakage)
split_point = int(len(X) * 0.8)
X_train, X_test = X.iloc[:split_point], X.iloc[split_point:]
y_train, y_test = y.iloc[:split_point], y.iloc[split_point:]

# Replace any inf/-inf with 0 (sometimes happens with pct_change)
X_train = X_train.replace([np.inf, -np.inf], 0)
X_test = X_test.replace([np.inf, -np.inf], 0)

# Train Model
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predictions
y_pred = rf_model.predict(X_test)

# ==========================================
# 5. Evaluation
# ==========================================
print("\nModel Evaluation Report:")
print(classification_report(y_test, y_pred))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred))

# Feature Importance
importances = rf_model.feature_importances_
feature_names = X.columns
plt.figure(figsize=(8, 4))
plt.barh(feature_names, importances)
plt.title("Feature Importance")
plt.xlabel("Importance")
plt.show()
```
